```python
#load libraries
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
#load dataset
df=pd.read_csv(r"E:/book.csv",encoding = "ISO-8859-1")
```

```python
#visualization of null 
sns.heatmap(df.isnull())
```

```python
#visualization
sns.countplot(x='BookRating',data=df)
```

```python
#changing the orientation of dataset accordingly
data=df.pivot_table(index='UserID',columns='BookTitle',values='BookRating').reset_index()
data
```

```python
#index resetting to unique userid
data.index=df.UserID.unique()
data=data.fillna(0)
data
```

```python
#cosine similarity
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine
```

```python
data2 = 1- pairwise_distances(data,metric='cosine')
data2
```

```python
data3=pd.DataFrame(data2)
np.fill_diagonal(data2, 0)
data3
```

```python
#relation between users
data3.index=df['UserID'].unique()
data3.columns=df['UserID'].unique()
```

```python
data3.iloc[0:5,0:5]
```

```python
#most similar users
data3.idxmax(axis=1)[0:50]
```
